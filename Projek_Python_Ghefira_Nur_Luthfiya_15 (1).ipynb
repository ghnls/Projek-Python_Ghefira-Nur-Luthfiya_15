```python
def konversi_panjang():
    print("\n=== Konversi Panjang ===")
    print("1. cm ke meter")
    print("2. meter ke cm")

    pilihan = input("Pilih (1/2): ")

    try:
        nilai = int(input("Masukkan nilai (harus bilangan bulat): "))
    except:
        print("❌ Input harus bilangan bulat (tidak boleh koma)!")
        return

    if pilihan == "1":
        hasil = nilai / 100
        print("Hasil:", hasil, "meter")
    elif pilihan == "2":
        hasil = nilai * 100
        print("Hasil:", hasil, "cm")
    else:
        print("Pilihan tidak valid")


def konversi_berat():
    print("\n=== Konversi Berat ===")
    print("1. gram ke kilogram")
    print("2. kilogram ke gram")

    pilihan = input("Pilih (1/2): ")

    try:
        nilai = int(input("Masukkan nilai (harus bilangan bulat): "))
    except:
        print("❌ Input harus bilangan bulat (tidak boleh koma)!")
        return

    if pilihan == "1":
        hasil = nilai / 1000
        print("Hasil:", hasil, "kg")
    elif pilihan == "2":
        hasil = nilai * 1000
        print("Hasil:", hasil, "gram")
    else:
        print("Pilihan tidak valid")


# Program utama
while True:
    print("\n=== APLIKASI KONVERSI ===")
    print("1. Konversi Panjang")
    print("2. Konversi Berat")
    print("3. Keluar")

    menu = input("Pilih menu (1-3): ")

    if menu == "1":
        konversi_panjang()
    elif menu == "2":
        konversi_berat()
    elif menu == "3":
        print("Terima kasih!")
        break
    else:
        print("Menu tidak tersedia")
```
